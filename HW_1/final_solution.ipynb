```python
# open files and set 'id' as an index
train_path = 'train.csv'
test_path = 'test.csv'
train = pd.read_csv(train_path)
test = pd.read_csv(test_path)
train = train.set_index('id')
test = test.set_index('id')
train.head(2)
```

```python
# let's mainly choose numeric values as out parameters
# convert date to three integers: day, month, year
# fill null values with mean or mode
train_x = train[['budget', 'popularity', 'runtime', 'release_date']]
train_x['runtime'].fillna(value=train_x['runtime'].mean(), inplace=True)
train_x['release_date'] = pd.to_datetime(train_x['release_date'])
train_x['release_y'] = train_x['release_date'].dt.year
train_x['release_m'] = train_x['release_date'].dt.month
train_x['release_d'] = train_x['release_date'].dt.day
train_x.drop(columns=['release_date'], inplace=True)
# feature scaling
train_x = (train_x - train_x.mean()) / train_x.std()
train_x['original_language'] = train['original_language']
train_x['status'] = train['status']
# make all of our values numeral
train_x = pd.get_dummies(train_x)
train_x.fillna(0)
train_x.info()
```

```python
# choose useful values for observation
test_x = test[['budget', 'popularity', 'runtime', 'release_date', 'original_language', 'status']]
# fill null values
test_x['release_date'].fillna(value='01-01-2000', inplace=True)
test_x.info()
```

```python
# function to compute cost, returns single integer
# characterized by the parameters theta
def computeCost(X, y, theta):
    inner = np.power(((X * theta.T) - y), 2)
    return np.sum(inner) / (2 * len(X))

# function to compute gradient, returns theta and cost
# performs gradient descent on the parameters theta using the update rules.
def gradientDescent(X, y, theta, alpha, iters):
    temp = np.matrix(np.zeros(theta.shape))
    parameters = int(theta.ravel().shape[1])
    cost = np.zeros(iters)
    
    for i in range(iters):
        error = (X * theta.T) - y
        
        for j in range(parameters):
            term = np.multiply(error, X[:,j])
            temp[0,j] = theta[0,j] - ((alpha / len(X)) * np.sum(term))
            
        theta = temp
        cost[i] = computeCost(X, y, theta)
        
    return theta, cost
```

```python
# convert values to matrix
train_x = np.matrix(train_x.values)
train_y = np.matrix(train_y.values)
theta = np.matrix(np.zeros(shape=(1,54)))
```
